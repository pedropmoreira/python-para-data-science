Coeficiente angular pode ser encontrado do seguinte modo :



a = coeficiente angular

y = Moscow

n = numero de elementos

X = datas

Captura de tela 2025-07-24 162656.png

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
(X**2).shape
```

```python
#calculando coeficiente angular
a = (n*np.sum(X*Y) - np.sum(X) * np.sum(Y) )/(n*np.sum(X**2)- np.sum(X)**2)
```

Coeficiente linear:Captura de tela 2025-07-24 163827.png

```python
b = np.mean(Y) - a * np.mean(X)
```

```python
y = a * X + b
```

```python
np.linalg.norm(Moscow-y)
```

```python
# calculando ponto especifico exemplo mes 41.5 podemos usar o que encontramos
plt.plot(datas,Moscow)
plt.plot(x,y)
plt.plot(41.5,41.5 * a +b , '*r')
# podemos estimar valores em meses futuros
plt.plot(100,100 * a +b , '*g')
```
